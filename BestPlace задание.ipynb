## Какова средняя выручка магазинов формата мини ТЦ за 2016 год?

```python
mini_tc = otlet_guide[otlet_guide['Формат магазина'].isin(['Мини ТЦ', 'Мини-ТЦ'])]
```

Уберем Nan значения в id, так как их нельзя объединить с другой таблицей

```python
mini_tc_ids = mini_tc['id точки'].dropna()
```

```python
# выбираем столбцы за 2016 год
cols_2016 = [i for i in revenue.columns if i.startswith('2016')]
```

Посмотрим на таблицу выручки, в ней видим, что есть пустые значения, уберем те, где вообще нет данных

```python
revenue_mini_tc = revenue[revenue['id точки'].isin(mini_tc_ids)][cols_2016]
revenue_mini_tc
```

```python
revenue_mini_tc = revenue_mini_tc.dropna(axis=0, how='all')
```

```python
print('Средняя выручка магазинов формата мини ТЦ за 2016 год составляет ' + str(round(revenue_mini_tc.mean(axis=1).mean(),2)))
```
