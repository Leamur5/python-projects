Сама функция и результат работы:

```python
def metrics_change(small_df,large_df):
    small_df = small_df.sort_values(by=['Class']) #сортируем строки, чтобы все считалось корректно
    large_df = large_df.sort_values(by=['Class'])
    
    presicion_growth = round((large_df['Presicion'] - small_df['Presicion']).mean(),3)
    presicion_proc = round(abs(((large_df['Presicion'] - small_df['Presicion'])/large_df['Presicion']).mean()*100),1)
    print('Presicion score вырос на ' + str(presicion_growth) + ' (это ' + str(presicion_proc) + ' %)')
    
    recall_growth = round((large_df['Recall'] - small_df['Recall']).mean(),3)
    recall_proc = round(abs(((large_df['Recall'] - small_df['Recall'])/large_df['Recall']).mean()*100),1)
    print('Recall score вырос на ' + str(recall_growth) + ' (это ' + str(recall_proc) + ' %)')
    
    map5_growth = round((large_df['mAP@.5'] - small_df['mAP@.5']).mean(),3)
    map5_proc = round(abs(((large_df['mAP@.5'] - small_df['mAP@.5'])/large_df['mAP@.5']).mean()*100),1)
    print('mAP@.5 score вырос на ' + str(map5_growth) + ' (это ' + str(map5_proc) + ' %)')
    
    map95_growth = round((large_df['mAP@.5:.95:'] - small_df['mAP@.5:.95:']).mean(),3)
    map95_proc = round(abs(((large_df['mAP@.5:.95:'] - small_df['mAP@.5:.95:'])/large_df['mAP@.5:.95:']).mean()*100),1)
    print('mAP@.5:.95: score вырос на ' + str(map95_growth) + ' (это ' + str(map95_proc) + ' %)')
    
```

```python
metrics_change(small_df,large_df)
```
